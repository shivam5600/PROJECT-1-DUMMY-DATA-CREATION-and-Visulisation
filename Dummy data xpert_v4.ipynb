### 1) Training wise

```python
#aggregating the scores by Sections (tra1_sec1 -- tra2_sec5)
traning_score = []
for t in [1,2]:
    traning_score.append("Traning_{}".format(t))

traning_score.insert(0,"user_name")
```

```python
#training wise data creation
tra1 = df_first_time.iloc[0:,1:361].sum(axis=1)
tra2 = df_first_time.iloc[0:,361:721].sum(axis=1)
training_wise_score = pd.DataFrame(columns=traning_score)
training_wise_score.user_name = df_first_time.user_name
training_wise_score.Traning_1 = tra1
training_wise_score.Traning_2 = tra2
total = training_wise_score.iloc[0:,1:].sum(axis=1)
training_wise_score['total'] = total
```

```python
training_wise_score.set_index('user_name',inplace= True)
```

```python
training_wise_score.head()
```

### 2) Section wise

```python
#aggregating the scores by Sections (tra1_sec1 -- tra2_sec5)
score_sections = []
for t in [1,2]:
    for s in [1,2,3,4]:
        score_sections.append("Tra{}_Section{}".format(t,s))

score_sections.insert(0,"user_name") 
```

```python
#selecting the desired data
a = df_first_time.iloc[0:,1:91].sum(axis=1)
b = df_first_time.iloc[0:,91:181].sum(axis=1)
c = df_first_time.iloc[0:,181:271].sum(axis=1)
d = df_first_time.iloc[0:,271:361].sum(axis=1)
e = df_first_time.iloc[0:,361:451].sum(axis=1)
f = df_first_time.iloc[0:,451:541].sum(axis=1)
g = df_first_time.iloc[0:,541:631].sum(axis=1)
h = df_first_time.iloc[0:,631:721].sum(axis=1)
section_wise_score = pd.DataFrame(columns=score_sections)
```

```python
## Putting data inside the columns
section_wise_score.user_name = df_first_time.user_name
section_wise_score.Tra1_Section1 = a
section_wise_score.Tra1_Section2 = b
section_wise_score.Tra1_Section3 = c
section_wise_score.Tra1_Section4 = d
section_wise_score.Tra2_Section1 = e
section_wise_score.Tra2_Section2 = f
section_wise_score.Tra2_Section3 = g
section_wise_score.Tra2_Section4 = h
total = section_wise_score.iloc[0:,1:].sum(axis=1)
section_wise_score['total'] = total
section_wise_score.set_index('user_name',inplace= True)
```

# Visualizing the data

```python
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as pyo
import matplotlib.pyplot as plt 
import dash
import dash_core_components as dcc
import dash_html_components as html
```

#### 1. Line chart for all 4 types of players (best, worst, improving, deteriorating) over time

```python
# Creating Line chart and exporting via DASH
n_df = section_wise_score.T
fig = go.Figure() 
fig.add_trace(go.Scatter(x=n_df.index[0:8],y=n_df['Dhanuk Subramanian'],
                    mode='lines+markers',
                    name='Top performer'))
fig.add_trace(go.Scatter(x=n_df.index[0:8],y=n_df['Madhav Sagar'],
                    mode='lines+markers',
                    name='worst performer'))
fig.add_trace(go.Scatter(x=n_df.index[0:8],y=n_df['Riaan Swamy'],
                    mode='lines+markers',
                    name='Improving performer'))
fig.add_trace(go.Scatter(x=n_df.index[0:8],y=n_df['Shayak Saran'],
                    mode='lines+markers',
                    name='Detoriating performer'))
fig.update_layout(title='Performers', xaxis_title='Sections', yaxis_title='Score_section_wise',height=720,width=1080)


app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False) 
```
