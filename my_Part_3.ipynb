### Selecting indivisual attempt columns and modifying it

```python
a1 = df.iloc[:,2:50]
```

```python
b1 = df.iloc[:,50:98]
```

```python
c1 = df.iloc[:,98:146]
```

```python
a1_sum = a1.sum(axis=1)
b1_sum = b1.sum(axis=1)
c1_sum = c1.sum(axis=1)
```

```python
a1b1c1 = a1_sum + b1_sum + c1_sum
```

## Creating a new dataframe for users vs scores in each attempts

```python
user_attmp = df.User_Name
```

```python
user_attmp = pd.DataFrame(user_attmp)
```

```python
#question in each attempt is 48

user_attmp['Attempt_1_score'] = a1_sum
user_attmp['Attempt_2_score'] = b1_sum
user_attmp['Attempt_3_score'] = c1_sum
user_attmp['Total_score'] = a1b1c1

```

```python
user_attmp.head()
```

## Now finding the Overall 'Top performers' and 'Bad performers'

```python
user_attmp.describe()
```

```python
best_perf_ = user_attmp[user_attmp['Total_score']>76].sort_values(by='Total_score',ascending=False)
best_perf_
```

```python
bad_perf_ = user_attmp[user_attmp['Total_score']<68].sort_values(by='Total_score',ascending=False)
bad_perf_
```

#### TOP PERFORMERS

```python
best_perf_.to_csv('top_25%_perf.csv')
```
